```python
numeric_cols = df.select_dtypes(include=['number']).columns
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
df[numeric_cols] = df[numeric_cols][~((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR)))]
df = df.dropna()
print("Outliers removed successfully!")
```

```python
from scipy import stats

# Selecting numerical columns for outlier detection
numerical_columns = ['danceability', 'energy', 'tempo']  # Add relevant numerical columns

# Calculating Z-scores for each numerical column
z_scores = stats.zscore(df[numerical_columns])
threshold = 2
# Finding outliers
outliers = (z_scores > threshold).any(axis=1)
# Displaying rows with outliers
print(df[outliers])
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder

# Load data from CSV
file_path = 'dataset.csv'
data = pd.read_csv(file_path)

# Separate features and target variable
X = data.drop('tempo', axis=1)
y = data['tempo']

# Convert string categorical variables to numerical using Label Encoding
label_encoder = LabelEncoder()
X_encoded = X.apply(label_encoder.fit_transform)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Initialize KNN regressor
k = 3
knn_regressor = KNeighborsRegressor(n_neighbors=k)

# Train the regressor
knn_regressor.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn_regressor.predict(X_test)
print("y_pred: ", y_pred)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
